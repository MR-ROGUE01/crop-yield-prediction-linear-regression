```python
import pandas as pd
import matplotlib.pyplot as plt
import joblib
```

```python
df_original = pd.read_csv(
    'crop_yield_cleaned_sample.csv'
)
```

```python
df_encoded = pd.get_dummies(
    df_original,
    columns=[
        'Region',
        'Soil_Type',
        'Crop',
        'Weather_Condition'
    ],
    drop_first=True
)

X = df_encoded.drop(
    'Yield_tons_per_hectare',
    axis=1
)
```

```python
plt.figure(figsize=(10,6))

df_original.boxplot(
    column='Yield_tons_per_hectare',
    by='Crop'
)

plt.title('Yield Distribution by Crop')
plt.suptitle('')
plt.xlabel('Crop Type')
plt.ylabel('Yield (tons/hectare)')
plt.show()
```

```python
region_yield = df_original.groupby(
    'Region'
)['Yield_tons_per_hectare'].mean()

plt.figure(figsize=(8,5))

region_yield.plot(
    kind='bar'
)

plt.title('Average Yield by Region')
plt.xlabel('Region')
plt.ylabel('Average Yield')
plt.show()
```
